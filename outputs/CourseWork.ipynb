Taska1

```python
import matplotlib.pyplot as plt
import pandas as pd
import json
import numpy as np

with open('transactions.txt', 'r') as file:
    jsonData = json.load(file)
    for jso in jsonData:
      jso[0]=jso[0].split('-')[1] + " " + jso[0].split('-')[0]
    print(jsonData[0])
    df = pd.DataFrame(jsonData, columns = ['Date','Number of transactions'])
    df = df.sort_values('Date')
    print(df)
    x=df['Date']
    y=df['Number of transactions']
    plt.figure(figsize = (20,10))
    plt.xticks(rotation=90)
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Number of transactions', fontsize=12)
    plt.bar(x,y,color='orange')
    
```

WashTrading

```python
import networkx as nx
import json as json
g = nx.DiGraph()
 

with open('wash.txt', 'r') as file:
    jsonData = json.load(file)
    for i in range(len(jsonData)):
      g.add_edge(jsonData[i][0][0], jsonData[i][0][1])
    
    #nx.draw(g)
    print(list(nx.recursive_simple_cycles(g)))
```

Gas guzzelers

```python
from matplotlib import dates
import matplotlib.pyplot as plt
import pandas as pd
import json
from decimal import Decimal
import numpy as np
with open('avgGasprice.txt', 'r') as file:
    jsonData = json.load(file)
    df = pd.DataFrame(jsonData, columns = ['Date','Gas price avg'])
    df["Date"] = pd.to_datetime(df["Date"])
    df=df.sort_values(by=['Date'])
    df['Gas price avg'] = df['Gas price avg'].astype('float')
    print(df.head())
```

```python
    formatter = dates.DateFormatter('%Y-%m')
    plt.figure(figsize = (10,5))
    ax = plt.gca()
    plt.xlabel('Month')
    plt.ylabel('Gas price avg')
    ax.xaxis.set_major_formatter(formatter)
    x=df['Date']
    y=df['Gas price avg']
    plt.plot(x,y,linewidth=3,color='orange')
    
```
